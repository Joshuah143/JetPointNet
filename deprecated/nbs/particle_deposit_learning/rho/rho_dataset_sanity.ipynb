```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import sys

# Assuming the necessary custom modules and functions (`pnet_part_seg_no_tnets`, `pnet_part_seg`) are available in the provided path
# For demonstration, we'll mock these functions

# Mocking custom model function `pnet_part_seg_no_tnets` for demonstration
def pnet_part_seg_no_tnets(input_points, input_channels, num_classes):
    input_tensor = keras.Input(shape=(input_points, input_channels))
    x = layers.Conv1D(64, 1, activation="relu")(input_tensor)
    x = layers.Conv1D(128, 1, activation="relu")(x)
    x = layers.Conv1D(1024, 1, activation="relu")(x)
    x = layers.GlobalMaxPooling1D()(x)
    x = layers.Dense(512, activation="relu")(x)
    x = layers.Dense(256, activation="relu")(x)
    output_tensor = layers.Dense(num_classes, activation="softmax")(x)

    model = keras.Model(inputs=input_tensor, outputs=output_tensor, name="mock_pnet_part_seg_no_tnets")
    return model

# Define the model using the mocked function
model = pnet_part_seg_no_tnets(1180, 5, 1)

# Using TensorFlow's function to plot the model
tf.keras.utils.plot_model(model, to_file='/mnt/data/model_plot.png', show_shapes=True, show_layer_names=True)

'/mnt/data/model_plot.png'
```
