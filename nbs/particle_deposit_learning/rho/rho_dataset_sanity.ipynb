```python

# Replace with your actual file path
file_path = '/data/mjovanovic/cell_particle_deposit_learning/rho/rho_processed_train_files/rho_len_6000_i_0.npz'

start_idx = 100
end_idx = start_idx + 50

# Load the .npz file
with np.load(file_path, allow_pickle=True) as data:
    X = data['X']
    Y = data['Y']
    
    # Print structure of arrays
    print("X array structure:")
    print(f"Shape: {X.shape}")
    print("Example items from the first sample of X:")
    # Printing the specified range of features of the first sample
    print(X[start_idx:end_idx, :5, :])  # Adjusted to print a range of samples

    print("\nY array structure:")
    print(f"Shape: {Y.shape}")
    print("Example items from the first sample of Y:")
    # Printing the specified range of targets of the first sample
    print(Y[start_idx:end_idx, :5, :])  # Adjusted to print a range of samples

    # Detailed view of a specific part of the dataset
    print("\nDetailed view of a specific sample's specific feature across all dimensions:")
    print(X[start_idx, start_idx, :])  # Specific sample, specific feature, all dimensions
    print("\nCorresponding target value for the specific feature of the specific sample:")
    print(Y[start_idx, start_idx, :])  # Corresponding target for the specific feature

    # Calculate and print the percentage of Y values that are strictly between 0 and 1, excluding -1
    valid_y_values = (Y > 0) & (Y < 1) & (Y != -1)
    valid_y_count = np.sum(valid_y_values)
    total_y_count = np.size(Y) - np.sum(Y == -1)  # Exclude -1 values from the total count
    percentage = (valid_y_count / total_y_count) * 100 if total_y_count > 0 else 0
    print(f"\nPercentage of Y values strictly between 0 and 1 (excluding -1): {percentage:.2f}%")
```
